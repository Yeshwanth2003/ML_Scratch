```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
X = data.data
y = data.target
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np

X = StandardScaler().fit_transform(X)
y = np.array(y)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y)
```

```python
class LogisticRegression:
    def __init__(self,lr,epoch):
        self.lr = lr
        self.epoch = epoch
        self.bias = 0
        self.weights = None
        
    def sigmoid(self,x):
        return 1/(1+np.exp(-x))
    
    def sigmoid_der(self,x):
           v = 1/(1+np.exp(-x)) 
           return v*(1-v)
       
    def fit(self,X,y):
        m,n = X.shape
        self.weights = np.random.uniform(-1,1,n)
        
        for i in range(self.epoch):
            pred = self.sigmoid(np.dot(X,self.weights)+self.bias)
            error = pred-y
            
            # dw = (1/m)*np.dot(X.T,error) #both are correct in programming perspective
            dw = (1/m)*np.dot(error,X)
            db = (1/m)*np.sum(error)

            self.weights-=self.lr*dw
            self.bias-=self.lr*db
    
    def predict(self,X):
        return [1 if i>0.5 else 0 for i in  self.sigmoid(np.dot(X,self.weights)+self.bias)]           
        
```

```python
model = LogisticRegression(0.01,5000)
model.fit(X_train,y_train)

pred = model.predict(X_test)
```
